```python
#Problem 1, all cases are shown in this code block:

#Import numpy, matplotlib.pyplot, and solve_ivp 
import numpy as np
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp

#General ODE: df/dt = a*f(t) + b(t)
def diff_eq(t, f, a, b_func):
    return  a*f + b_func(t)

#Defining the different cases
cases = [
    {"a": -2, "b": lambda t: 0, "title": "a = -2, b(t) = 0"},
    {"a": -2j, "b": lambda t: 0, "title": "a = -2i, b(t) = 0"},
    {"a": -1, "b": lambda t: np.cos(t/3), "title": "a = -1, b(t) = cos(t/3)"},
    {"a": -1, "b": lambda t: np.exp(t), "title": "a = -1, b(t) = exp(t)"}
]

#Time span for integration (I picked 0–5 because it’s long enough to show the decay and oscillations in the first three cases, but not so long that the exponential forcing in the last case blows up and makes the plot unreadable.)
t_span = (0, 5)
t_eval = np.linspace(0, 5, 100) 
f0 = [1] #Initial condition f(0) = 1

#Creating figure for plotting 
plt.figure(figsize=(12,10))

#Solving and plotting each case
for i, case in enumerate(cases, 1):
    sol = solve_ivp(diff_eq, t_span, f0, args=(case ["a"], case["b"]), t_eval=t_eval)

#Creating subplot
    plt.subplot(2,2,i)

#Handling the complex case
    if case["a"] == -2j:
        plt.plot(sol.t, np.real(sol.y[0]), label = 'Real part')
        plt.plot(sol.t, np.imag(sol.y[0]), label = 'Imaginary part')
        plt.legend()
    else:
        plt.plot(sol.t, sol.y[0])
        
    plt.title(case["title"])
    plt.xlabel('Time')
    plt.ylabel('f(t)')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Problem 2:

#Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

#Defining the system of differential equations
def system_eq(t, variables):
    """
    System of equations:
    dx/dt + 3y = 0 -> dx/dt = -3y
    dy/dt - x = 0 -> dy/dt = x
    """
    x, y = variables
    dxdt = -3*y
    dydt = x
    return [dxdt, dydt]

#Initial condition: x(0) = 2, y(0) = -1
initial_conditions = [2, -1]

#I chose 0–10 because it shows multiple full oscillations in both x(t) and y(t), making the periodic behavior and the elliptical phase portrait clear.
t_span = (0, 10)
t_eval = np.linspace(0, 10, 500) 

#Solving the system
sol = solve_ivp(system_eq, t_span, initial_conditions, t_eval=t_eval)

#Creating plot
plt.figure(figsize=(12, 5))

#Plotting x(t) and y(t) seperately
plt.subplot(1,2,1)
plt.plot(sol.t, sol.y[0], label='x(t)', linewidth=2)
plt.plot(sol.t, sol.y[1], label= 'y(t)', linewidth=2)
plt.title('Time Evolution of x(t) and y(t)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

#Phase portrait plot (x vs. y)
plt.subplot(1,2,2)
plt.plot(sol.y[0], sol.y[1], linewidth=2)
plt.title('Phase Portrait: x vs. y')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.grid(True)
plt.axis('equal') #Equal scaling for proper ellipse shape

plt.tight_layout()
plt.show()
```

```python
#Problem 3

#Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

#Defining the 3 variable system of differential equations
def system_eq(t, variables):
    """
    System of equations:
    dx/dt = -2x - 2y - 4z
    dy/dt = -2x + y - 2z
    dz/dt = 5x + 2y + 7z
    """
    x, y, z = variables
    dxdt = -1*x - 2*y - 4*z
    dydt = -2*x + y - 2*z
    dzdt = 5*x + 2*y + 7*z
    return [dxdt, dydt, dzdt]

#Initial conditions weren't given so I used 1,1,1
initial_conditions = [1,1,1]

#I chose 0–5 because it’s long enough to clearly show how the three variables evolve and interact, but not so long that the fast-growing terms dominate and make the plots hard to read.
t_span = (0, 5)
t_eval = np.linspace(0,5,500)

#Solving the system
sol = solve_ivp(system_eq, t_span, initial_conditions, t_eval=t_eval)

#Creating the plot
plt.figure(figsize=(15,5))

#Plotting all three variables over time
plt.subplot(1,2,1)
plt.plot(sol.t, sol.y[0], label = 'x(t)', linewidth=2)
plt.plot(sol.t, sol.y[1], label = 'y(t)', linewidth=2)
plt.plot(sol.t, sol.y[2], label = 'z(t)', linewidth=2)
plt.title('Time Evolution of x(t), y(t), z(t)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Additional 2D phase portraits for different perspectives
plt.figure(figsize=(15, 5))

# x vs y
plt.subplot(1, 3, 1)
plt.plot(sol.y[0], sol.y[1], linewidth=2)
plt.title('2D Phase Portrait: x vs y')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.grid(True)

# x vs z
plt.subplot(1, 3, 2)
plt.plot(sol.y[0], sol.y[2], linewidth=2)
plt.title('2D Phase Portrait: x vs z')
plt.xlabel('x(t)')
plt.ylabel('z(t)')
plt.grid(True)

# y vs z
plt.subplot(1, 3, 3)
plt.plot(sol.y[1], sol.y[2], linewidth=2)
plt.title('2D Phase Portrait: y vs z')
plt.xlabel('y(t)')
plt.ylabel('z(t)')
plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
#Problem 4

#Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

#Defining the 3 variable system of differential equations
def system_eq(t, variables):
    """
    System of equations:
    dx/dt = 4x - y + (1 - e^{-3t})(t + sin(t))
    dy/dt = x + 2y + t(1 - e^{3t})cos(t) 
    """
    x, y = variables
    exp_term = 1 - np.exp(-3*t)

    dxdt = 4*x - y + exp_term*(t + np.sin(t))
    dydt = x + 2*y + t*exp_term*np.cos(t)

    return [dxdt, dydt]

#Initial conditions: x(0) = 0, y(0) = 1
initial_conditions = [0, 1]

#I picked 0–10 because it’s long enough to capture how the forcing terms and exponential factors influence the system, 
showing both the transient behavior at the start and the longer-term growth of x(t) and y(t) 
t_span = (0, 10)
t_eval = np.linspace(0,10,500)

#solving the system
sol = solve_ivp(system_eq, t_span, initial_conditions, t_eval=t_eval)

#Creating the plot
plt.subplot(1, 2, 1)
plt.plot(sol.t, sol.y[0], label='x(t)', linewidth=2)
plt.plot(sol.t, sol.y[1], label='y(t)', linewidth=2)
plt.title('Time Evolution of x(t) and y(t)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

# Ploting phase portrait (x vs y)
plt.subplot(1, 2, 2)
plt.plot(sol.y[0], sol.y[1], linewidth=2)
plt.title('Phase Portrait: x vs y')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
